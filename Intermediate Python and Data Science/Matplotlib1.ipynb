### Use shorthand notation
**Format Strings**

A format string consists of a part for color, marker and line:

`fmt = '[marker][line][color]'`

Each of them is optional. If not provided, the value from the style cycle is used. Exception: If line is given, but no marker, the data will be a line without markers.

Other combinations such as [color][marker][line] are also supported, but note that their parsing may be ambiguous.

**Check the Matplotlib documentation to view all available:**
 - Markers (e.g. 'o', 's', '*', etc.)
 - Line styles (e.g. '-', '--', '-.', ':')
 - Colors (e.g. 'r', 'g', 'b', 'k', etc.)

```python
# Simple line plot: The first list is X-values, and the second list is Y-values.
x = [0,1,2,3,4]
y = [0,2,4,6,8]


# shorthand notation
plt.plot(x,y,'b.--',label='2x')
# Or plt.plot([1,2,3],[4,5,6])
# Note: Use linestyle='--' to change the line style.

# title: Set a title for the Axes.
plt.title('First Graph', fontdict={'fontname':'Humor Sans','fontsize':20})

# xlabel: Set the label for the x-axis.
plt.xlabel('X Axis')
# ylabel: Set the label for the y-axis.
plt.ylabel('Y Axis')

# Set custom tick values for the X-axis
plt.xticks([0,1,2,3,4])
# Set custom tick values for the y-axis
plt.yticks([0,2,4,6,8,10])

# Add a legend to the plot. It will display the label specified in plt.plot(..., label='...') above.
plt.legend()



# To remove this line => [<matplotlib.lines.Line2D at 0x7e828487c440>] => Use plt.show()
plt.show()# Showing the chart clearly

```

### Third Graph
These two lines plot the function x² in two separate parts using different colors and line styles.

- x2[:5] takes the first half of the data and plots it with a solid cyan line ('c').

- x2[4:] takes the remaining values and plots them with a magenta dashed line ('m--').

This shows how you can split data and use different styles to highlight sections of a curve.

```python
# Simple line plot: The first list is X-values, and the second list is Y-values.
x = [0,1,2,3,4]
y = [0,2,4,6,8]


# shorthand notation
plt.plot(x,y,'b.--',label='2x')

# title: Set a title for the Axes.
plt.title('Third Graph', fontdict={'fontname':'Humor Sans','fontsize':20})

# xlabel: Set the label for the x-axis.
plt.xlabel('X Axis')
# ylabel: Set the label for the y-axis.
plt.ylabel('Y Axis')




x2 = np.arange(0,4.5,0.5)
# Split the curve into two parts with different colors and line styles.
plt.plot(x2[:6],x2[:6]**2,'c',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'m--',label='x^2')


# Add a legend to the plot. It will display the label specified in plt.plot(..., label='...') above.
plt.legend()



plt.show()
```

### Fourth Graph
**Resize Graph**

 plt.figure(figsize=(5,3), dpi=100) resizes the plot to 5×3 inches

 with a resolution of 100 DPI, resulting in a 500×300 pixel image.

```python
# Simple line plot: The first list is X-values, and the second list is Y-values.
x = [0,1,2,3,4]
y = [0,2,4,6,8]


# Resize Graph (dpi specifies pixels per insh. When saving probably should use 300 if possible)
plt.figure(figsize=(5,3),dpi=100)

# Line 1

# shorthand notation
plt.plot(x,y,'b.--',label='2x')

# title: Set a title for the Axes.
plt.title('Fourth Graph', fontdict={'fontname':'Humor Sans','fontsize':15})


# X and Y Labels
# xlabel: Set the label for the x-axis.
plt.xlabel('X Axis')
# ylabel: Set the label for the y-axis.
plt.ylabel('Y Axis')



# Line 2

# select interval we want to plot points at.
x2 = np.arange(0,4.5,0.5)

# Split the curve into two parts with different colors and line styles.
plt.plot(x2[:6],x2[:6]**2,'c',label='x^2') # Plot part of the graph as line.
plt.plot(x2[5:],x2[5:]**2,'m--',label='x^2') # Plot remainder og graph as a dot.


# Add a legend to the plot.
plt.legend()

# Save the current figure as a PNG image (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png',dpi=300)

# Show plot
plt.show()
```

### Bar Chart

A bar chart is used to compare different categories or groups using rectangular bars.

Each bar’s height (or length) represents the value of that category, making it easy to visualize differences and trends among discrete data points.

```python
# Define category labels and their corresponding values.
labels = ['A','B','C']
values = [1,4,2]

# Create a bar chart.
plt.bar(labels,values)

plt.figure(figsize=(6,4))
plt.show()
```

### Add hatches
hatches = ['/', '\\', '|', '-', '+', 'x', 'o', 'O', '.', '*']


https://matplotlib.org/stable/gallery/shapes_and_collections/hatch_style_reference.html

```python
labels =['A','B','C']
values = [1,4,2]

bars = plt.bar(labels,values)

# Apply a hatch pattern (diagonal lines) to the first bar for visual distinction.
bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('+')

plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels,values)

patterns = ['/','O','+']
for bar in bars:
  bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(6,4))

plt.show()
```
