### Task 1: Data Cleaning
Create a Pandas DataFrame from a dictionary with columns `customer_id` (integers), `purchase_amount` (floats, some missing), and `category` (strings, some duplicates). Include at least 5 rows, with at least 1 missing value and 1 duplicate row.

- Fill missing `purchase_amount` values with the column‚Äôs mean.
- Remove duplicate rows based on all columns.
- Display the cleaned DataFrame.


```python
import pandas as pd
data = {
    'customer_id':[1,2,3,4,2],
    'purchase_amount': [100.0, None, 150.0, 200.0, 100.0],
    'category': ['Electronics', 'Books', 'Electronics', 'Clothing', 'Books']
}
df1 = pd.DataFrame(data)
display(df1)
```

```python
df1['purchase_amount'] = df1['purchase_amount'].fillna(df1['purchase_amount'].mean())
df1 = df1.drop_duplicates()
display(df1)
```

### Task 2: Grouping and Aggregation
Using the cleaned DataFrame from Task 1, group by `category` and calculate:
- The total `purchase_amount` for each category.
- The average `purchase_amount` for each category.
- Store results in a new DataFrame with columns `category`, `total_amount`, and `avg_amount`.
- Display the aggregated DataFrame.

```python
grouped = df1.groupby('category')['purchase_amount'].agg(['sum','mean']).reset_index()
grouped.columns = ['category','total_sum','avg_amount']
display(grouped)
```

### Task 3: Merging DataFrames
Create a second DataFrame with columns `customer_id` (integers, matching some IDs from Task 1) and `region` (strings) with at least 3 rows. Merge it with the cleaned DataFrame from Task 1 using an **inner merge** on `customer_id`. Display the merged DataFrame.

```python
data2 = {
    'customer_id':[1,2,5],
    'region':['North','South','East'],
}
df_region = pd.DataFrame(data2)
merged_df = pd.merge(df1,df_region,on='customer_id',how='inner')
display(merged_df)
```
